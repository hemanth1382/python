```python
import matplotlib.pyplot as plt

time = []
for i in range(8):
    time.append(i)
position = [1.86,0.83,0.36,0.29,0.27,0.32,0.31,3.15]
plt.rcParams['figure.figsize'] = [5,15]

plt.plot(time, position)
plt.xlabel("log(block_size)/log2")
plt.ylabel('L1 miss rate')
plt.savefig('plot128.png', dpi=300, bbox_inches='tight')
```
